# **Statistics Advance Part 1 Module 02nd. Questions**


**1. What is a random variable in probability theory?**

A random variable is a variable that takes numerical values based on the outcome of a random experiment. It helps us measure and analyze uncertainty.

**2. What are the types of random variables?**

There are two types: discrete (countable values like number of students) and continuous (measurable values like height or weight).

**3. What is the difference between discrete and continuous distributions?**

Discrete distributions deal with specific values (like 0,1,2), while continuous distributions deal with any value in a range (like 0.5, 1.3, etc.).

**4. What are probability distribution functions (PDF)?**

PDFs describe the likelihood of different outcomes for continuous random variables. The area under the curve shows probabilities.

**5. How do cumulative distribution functions (CDF) differ from PDFs?**

CDF gives the probability that a value is less than or equal to a certain point, while PDF shows the likelihood at each exact value.

**6. What is a discrete uniform distribution?**

It’s a type of distribution where all outcomes are equally likely. Example: rolling a fair die (each number 1–6 has equal chance).

**7. What are the key properties of a Bernoulli distribution?**

Bernoulli distribution has only two outcomes: success (1) and failure (0), like flipping a coin once.

**8. What is the binomial distribution, and how is it used in probability?**

It models the number of successes in a fixed number of independent Bernoulli trials. Example: 3 heads in 5 coin tosses.

**9. What is the Poisson distribution and where is it applied?**

It models the number of events occurring in a fixed interval of time or space. Common in modeling calls per minute in a call center.

**10. What is a continuous uniform distribution?**

It represents equal probability for all values in a given range. Example: picking a number randomly between 0 and 1.

**11. What are the characteristics of a normal distribution?**

It’s symmetric, bell-shaped, and most data clusters around the mean. Mean = Median = Mode in a perfect normal curve.

**12. What is the standard normal distribution, and why is it important?**

It’s a normal distribution with mean 0 and standard deviation 1. It simplifies calculations and is the basis for Z-scores.

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

CLT says that the sampling distribution of the sample mean approaches normality as sample size increases, even if data isn't normal.

**14. How does the Central Limit Theorem relate to the normal distribution?**

It allows us to use normal distribution to make inferences about population means, especially for large samples.

**15. What is the application of Z statistics in hypothesis testing?**

Z-statistics help test hypotheses by comparing sample data to population parameters using standard normal distribution.

**16. How do you calculate a Z-score, and what does it represent?**

Z = (X - μ) / σ. It shows how many standard deviations a value is from the mean. Useful for comparing data points.

**17. What are point estimates and interval estimates in statistics?**

Point estimate gives a single value (like sample mean). Interval estimate gives a range (confidence interval) likely to contain the true value.

**18. What is the significance of confidence intervals in statistical analysis?**

They give a range where the true population value likely falls, offering more reliable insights than a single point estimate.

**19. What is the relationship between a Z-score and a confidence interval?**

Z-scores help build confidence intervals by defining the number of standard deviations around the mean to cover.

**20. How are Z-scores used to compare different distributions?**

Z-scores standardize different datasets, so you can compare values on the same scale, even with different units or spreads.

**21. What are the assumptions for applying the Central Limit Theorem?**

Sample size should be large (usually n ≥ 30), and observations should be independent and identically distributed.

**22. What is the concept of expected value in a probability distribution?**

Expected value is the average result of a random variable over many trials. It's like a long-term prediction.

**23. How does a probability distribution relate to the expected outcome of a random variable?**

It defines the possible outcomes and their probabilities, helping us calculate the expected (average) result.
